```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Suponiendo que ya tienes un DataFrame "df" con columnas:
# [ "AÑO", "cod_comuna", "edad", "tipo_edad", ... ]

# 1) Renombra "AÑO" a "anio" (opcional, pero recomendado para consistencia)
df.rename(columns={'AÑO': 'anio'}, inplace=True)

# 2) Convierte cod_comuna a variable categórica (porque es un ID)
df['cod_comuna'] = df['cod_comuna'].astype('category')

# 3) Crea una nueva columna con la edad en años
def convertir_edad_flexible(row):
    """
    tipo_edad = 1 => edad en años
    tipo_edad = 2 => edad en meses (dividimos entre 12)
    tipo_edad = 3 => edad en días (dividimos entre 365)
    """
    if row['tipo_edad'] == 1:    # años
        return row['edad']
    elif row['tipo_edad'] == 2: # meses
        return row['edad'] / 12
    elif row['tipo_edad'] == 3: # días
        return row['edad'] / 365
    else:
        return np.nan

df['edad_en_anios'] = df.apply(convertir_edad_flexible, axis=1)

# 4) Selecciona solo las columnas numéricas verdaderamente relevantes
#    (ajusta según tus columnas útiles para el análisis)
df_numeric = df[['anio', 'edad_en_anios']]

# 5) Calcula la matriz de correlación
corr_matrix = df_numeric.corr()

# 6) Genera el heatmap
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
plt.title("Matriz de correlación de variables numéricas")
plt.show()


```
